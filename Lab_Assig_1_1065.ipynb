```python

# Student ID: 65

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1: Generate Dataset

k = 65
np.random.seed(65)
data = 60 + np.random.randint(0, 15, size=12) + (k % 5)

print("Generated Dataset (Exam Scores):", data)


# 2: State Hypotheses
mu0 = 65


# 3: Perform One-Sample t-Test

t_stat, p_value = stats.ttest_1samp(data, mu0)
df = len(data) - 1
sample_mean = np.mean(data)

print("\n--- One-Sample t-Test Results ---")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Hypothesized Mean: {mu0}")
print(f"t-Statistic: {t_stat:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"p-Value: {p_value:.4f}")

# 4: 95% Confidence Interval

confidence = 0.95
ci = stats.t.interval(confidence, df, loc=sample_mean, scale=stats.sem(data))

print("\n95% Confidence Interval for the Mean:", ci)


# Step 5: Decision (Î± = 0.05)
alpha = 0.05
if p_value < alpha:
    decision = "Reject H0 â†’ The mean is significantly different from 65."
else:
    decision = "Fail to Reject H0 â†’ No significant difference from 65."

print("\nDecision:", decision)


# 6: Visualization

plt.figure(figsize=(7,5))
plt.errorbar(1, sample_mean, yerr=stats.sem(data), fmt='o', capsize=5, label="Sample Mean Â± SE")
plt.axhline(mu0, color='r', linestyle='--', label="Hypothesized Mean (65)")
plt.xlim(0, 2)
plt.xticks([])
plt.ylabel("Exam Scores")
plt.title("One-Sample t-Test Visualization")
plt.legend()
plt.show()
```
