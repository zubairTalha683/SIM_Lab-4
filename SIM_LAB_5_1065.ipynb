```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Define the dataset (observed sample)
data = np.array([5.2, 5.4, 5.1, 5.6, 5.3, 5.8])

# Step 2: Define the population mean (hypothesized mean)
population_mean = 5.5  # Null Hypothesis (Hâ‚€)

# Step 3: Perform the one-sample t-test
t_statistic, p_value = stats.ttest_1samp(data, population_mean)

# Step 4: Calculate summary stats
sample_mean = np.mean(data)
df = len(data) - 1
ci = stats.t.interval(0.95, df=df, loc=sample_mean, scale=stats.sem(data))

# Step 5: Display the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Hypothesized Population Mean: {population_mean:.2f}")
print(f"t-Statistic: {t_statistic:.2f}")
print(f"Degrees of Freedom: {df}")
print(f"p-Value: {p_value:.4f}")
print(f"95% Confidence Interval: [{ci[0]:.2f}, {ci[1]:.2f}]")

# Step 6: Conclusion
alpha = 0.05
if p_value < alpha:
    print(f"Reject Hâ‚€: The sample mean is significantly different from {population_mean}.")
else:
    print(f"Fail to reject Hâ‚€: The sample mean is not significantly different from {population_mean}.")

# Step 7: Visualization
plt.figure(figsize=(6,4))
plt.bar(1, sample_mean, color="skyblue", label="Sample Mean")
plt.errorbar(1, sample_mean, yerr=np.std(data, ddof=1), fmt='k', linewidth=1.5)
plt.axhline(y=population_mean, color='r', linestyle='--', linewidth=1.5, label='Hypothesized Mean')
plt.title("Sample Mean vs Hypothesized Mean")
plt.xlim(0.5,1.5)
plt.xticks([])
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```
